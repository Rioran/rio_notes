```python
from rionotes import Note
```

Create a chord by combining several notes together:

```python
note = Note('c2') + Note('c3') + Note('e4') + Note('g4')
note.display()
```

See a small portion of the chord sound wave:

```python
note.plot()
```

Or a full soundwave:

```python
note.plot(full=True)
```

Apply ADSR technique for the sound to be more alive:

```python
note.apply_adsr()
note.display()
note.plot(full=True)
```

Change sound wave form:

```python
Note.wave_type = "triangular"
note = Note('c2') + Note('c3') + Note('e4') + Note('g4')
note.apply_adsr()
note.display()
note.plot(full=True)
```

Smooth things a bit to acquire a less aggressive sound:

```python
note.smooth(wing=10, times=5)
note.display()
note.plot(full=True)
```

Save for the result to be used elsewhere:

```python
print(note.name)  # future filename
```

```python
note.save()
```
